```python
import ipaddress 

#converting IPv4 address to int 

def ip2int(ip):
  if ip=='0' or ip=='' or ip=='0.0' or ip==0.0 or ip==1.0 or ip==2.0:
        return 0
  else:
        return int(ipaddress.ip_address(ip))
#for ind in range(100000):
  #df['ip.dst_host'][ind]=ip2int(df['ip.dst_host'][ind])
```

```python

from sklearn.preprocessing import LabelEncoder
 
#Create the function to fit and transform the label encoder
def label_encode_columns(df, columns):
    encoders = {}
    for col in columns:
        le = LabelEncoder().fit(df[col].astype('str'))
        df[col] = le.transform(df[col].astype('str'))
        encoders[col] = le
    return df, encoders

 
 
#Define the columns we want to encode
encode_columns=[
"tcp.payload",
"http.request.version",
"http.request.full_uri",
"http.referer",
"http.request.method",
"http.file_data",
"http.request.uri.query",
"mqtt.msg",
"mqtt.conack.flags",
"mqtt.topic",
"mqtt.protoname",
"dns.qry.name.len",
"tcp.options",
"tcp.srcport"
     ]
 
#Fit and transform the training dataset, returing both the new training dataset and the fitted encoders to use on the scoring dataset
df, encoders = label_encode_columns(df=df, columns=encode_columns)
df
```
